# MNIST-like datasets

I would like to prepare the datasets in a uniform way and compute various targets once, so I reject usual torchvision dataloaders.

```python
def save_to_folder(dataloader, datafolder):
    if not op.exists(datafolder):
        os.makedirs(datafolder)
    for i,a in tqdm(enumerate(dataloader)):
        a[0].save(op.join(datafolder, str(i)+"_"+str(a[1])+".png"))
```

## MNIST

```python
mnist_train = datasets.MNIST(root="../raw", train=True, download=True)
```

```python
mnist_test = datasets.MNIST(root="../raw", train=False, download=True)
```

```python
save_to_folder(mnist_train, "../data/MNIST_train")
```

```python
save_to_folder(mnist_test, "../data/MNIST_test")
```
